```python
from wordcloud import WordCloud
from scipy import signal
cities = data.city.dropna()
plt.subplots(figsize=(10,10))
wordcloud = WordCloud(background_color = 'white',
                     width = 512,
                     height = 384).generate(' '.join(cities))
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```

```python
x_year = data['year'].unique()
y_count_years = data['year'].value_counts().sort_index()
plt.figure(figsize = (18,10))
sb.barplot(x = x_year,
           y = y_count_years,
           palette = 'rocket')
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year')
plt.title('Attack_of_Years')
plt.show()
```

```python
pd.crosstab(data.attacktype, data.city).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Attacks by cities')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
pd.crosstab(data.year, data.region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```
